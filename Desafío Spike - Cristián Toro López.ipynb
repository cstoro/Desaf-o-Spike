## 4.-Casos Extremos

Crea tres variables llamadas 

● caudal_extremo 

● temp_extremo  

● precip_extremo

Dichas variables toman valor 1 un día si el caudal/temperatura/precipitación (según sea el caso)  observado ese día es extremo. Esto significa que es mayor de lo "esperado". Para capturar esta idea,  el valor de caudal, por ejemplo, toma valor 1 si está sobre el percentil 95 de ese caudal para esa  estación del año (Verano, Primavera, Otoño, Invierno). Toma valor 0 cuando está bajo ese percentil.  En otras palabras, para cada estación de medición y para cada estación del año, debes considerar la  distribución histórica de caudal/temperatura/precipitación para elegir ese percentil 95. 
Esta medida toma en cuenta la estacionalidad, pues, por ejemplo, una temperatura de 25 grados en  invierno puede ser extrema, pero en verano es normal. También toma en cuenta que cada cuenca (o  estación) es diferente. Lo que es extremo para una cuenca no lo es para la otra. 
¿Les parece razonable esta medida para capturar algo “extremo”? ¿Usarían otra? ¿Cuál? (​Solamente  descríbanla, no la codifiquen! Vamos a usar la definición de Spike para esta desafío) 


```python
import pandas as pd
import numpy as np

df2=df[['codigo_estacion','codigo_cuenca','altura','latitud','longitud','nombre','fecha','caudal','gauge_id','precip_promedio','temp_max_promedio']] 
        #me quedo solo con las columnas imxtantes por ahora.        
df2.fecha=pd.to_datetime(df2.fecha)
#print(df2.fecha.dt.month)
df2['estacion']=df2.fecha.dt.month
#por simplicidad, se consideraran las estaciones por meses, además dado el cambio climatico ultimamente 10 dias no marcan diferencias en 
#cuanto a condiciones de clima.
dkf=[]
for i in df2['estacion']:
    if i in [1,2,3]:
        dkf.append('verano')
    elif i in [4,5,6]:
        dkf.append('otoño')
    elif i in [7,8,9]:
        dkf.append('invierno')
    else:
        dkf.append('primavera')

        
df2['estacion']=pd.Series(dkf)
df2['estacion'].head()

codigo_estacion_q=[]
estacion_q=[]
q95_caudal=[]
q95_temp=[]
q95_precip=[]

for indice,fila in df2.drop_duplicates(['codigo_estacion','estacion'])[['codigo_estacion','estacion']].iterrows():
    codigo_estacion_q.append(fila['codigo_estacion'])
    estacion_q.append(fila['estacion'])
    q95_caudal.append(df2[(df2['codigo_estacion']==fila['codigo_estacion'])&(df2['estacion']==fila['estacion'])]['caudal'].quantile(0.95))
    q95_temp.append(df2[(df2['codigo_estacion']==fila['codigo_estacion'])&(df2['estacion']==fila['estacion'])]['temp_max_promedio'].quantile(0.95))
    q95_precip.append(df2[(df2['codigo_estacion']==fila['codigo_estacion'])&(df2['estacion']==fila['estacion'])]['precip_promedio'].quantile(0.95))

df_q95=pd.DataFrame({'codigo_estacion':codigo_estacion_q,'estacion':estacion_q,'q95_caudal':q95_caudal,'q95_temp':q95_temp,'q95_precip':q95_precip})
#ya tengo ahora un dataframe con los q955 de las variables solicitadas, ahora iteraremos y crearemos las variables del enunciado

df_3=pd.merge(df2, df_q95, on=['codigo_estacion','estacion'], how='left')

df_3['caudal_extremo']=np.where(df_3['caudal']>df_3['q95_caudal'],1,0)
df_3['temp_extremo']=np.where(df_3['temp_max_promedio']>df_3['q95_temp'],1,0)
df_3['precip_extremo']=np.where(df_3['precip_promedio']>df_3['q95_precip'],1,0)

df_3.head() #este dataframe contiene todo lo solicitado y algunas columnas extra.(los percentiles)

```

En cuanto a la metodologia de calculo de los casos extremos, considero correcto hacer una diferenciaciòn por estacion.
